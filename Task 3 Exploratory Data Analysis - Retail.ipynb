# EDA

# Interpreting Maximum Profit on the basis of Category

```python
plt.figure(figsize=(12,8))
sns.barplot(x='Category',y='Profit',data=df)
```

# Interpreting Maximum Profit on the basis of Sub-Category

```python
plt.figure(figsize=(12,12))
sns.barplot(x='Sub-Category',y='Profit',data=df)
plt.tight_layout
plt.xticks(rotation=90)
```

# Interpreting Maximum Profit on the basis of Region

```python
plt.figure(figsize=(12,8))
sns.barplot(x='Region',y='Profit',data=df)
```

# Interpreting Maximum Sales on the basis of Region

```python
plt.figure(figsize=(12,8))
sns.barplot(x='Region',y='Sales',data=df)
```

# Interpreting Maximum Sales on the Basis of Category 

```python
plt.figure(figsize=(12,8))
sns.barplot(x='Category',y='Sales',data=df)
```

# Interpreting Maximum Sales on the basis of Sub Category

```python
plt.figure(figsize=(12,12))
sns.barplot(x='Sub-Category',y='Sales',data=df)
plt.tight_layout
plt.xticks(rotation=90)
```

# Interpreting the Region getting Maximum Discount

```python
plt.figure(figsize=(8,6))
sns.barplot(x='Region',y='Discount',data=df)
```

# Interpreting the Segment getting Maximum Discount

```python
plt.figure(figsize=(8,6))
sns.barplot(x='Segment',y='Discount',data=df)
```

# Interpreting the Category getting Maximum Discount

```python
plt.figure(figsize=(8,6))
sns.barplot(x='Category',y='Discount',data=df)
```

# Interpreting the Sub-Category getting Maximum Discount

```python
plt.figure(figsize=(12,12))
sns.barplot(x='Sub-Category',y='Discount',data=df)
plt.tight_layout
plt.xticks(rotation=90)
```
